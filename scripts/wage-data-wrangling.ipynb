```python
# Wrangles monthly SW MT nonmetro area wage estimates into a cleaned .json
```

```python
import pandas as pd
```

```python
pd.set_option('max_columns', None)
pd.set_option('max_rows', 50)
```

```python
# Input files
# paths relative to scripts folder

raw_excel_file = './../raw-data/oesm15ma/BOS_M2015_dl.xlsx'
column_keys_file = './../raw-data/column_keys.json'
note_keys_file = './../raw-data/note_keys.json'

# column_keys.json and note_keys.json 
# generated manually from oesm15ma/field_descriptions.xlsx

na_values = ['*','**','#','~']

# CONSTANTS

area_code = 3000003
area_name = 'Southwest Montana'
```

```python
wages = pd.read_excel(raw_excel_file, na_values=na_values)
```

```python
# filter to SW Montana
wages = wages[wages['AREA'] == area_code]
print(len(wages))
```

```python
# Professions with only annual wages reported
wages[wages['ANNUAL'] == True]['OCC_TITLE']
```

```python
# Professions with only hourly wages reported
wages['HOURLY']
```

```python
# Focus to columns of interest
# Focusing on annual wages because all professions have them
# Neglecting averages here
include_cols = ['OCC_CODE', 'OCC_TITLE', 'OCC_GROUP', 'TOT_EMP',
                'EMP_PRSE', 'A_PCT10', 'A_PCT25', 'A_MEDIAN',
                'A_PCT75', 'A_PCT90']
```

```python
cleaned = wages[include_cols]
```

```python
cleaned
```

```python
categories = cleaned[cleaned['OCC_GROUP'] == 'major']
```

```python
categories
```

```python
categories.to_json('./../data/categories.json', orient='records')
```
